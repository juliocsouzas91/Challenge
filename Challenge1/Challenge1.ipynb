# Problem 2

```python
def CheckNewSolution(NewBuf,IntSeq):
    Accepted = 0
    Solution = NewBuf[0]
    Sum = int(NewBuf[0])
    for aux1 in range(1,len(NewBuf)):
        Solution += NewBuf[aux1]
        Sum = Sum + int(NewBuf[aux1])
    Solution = int(Solution)
    if(Sum==len(IntSeq)+1):
        Accepted = 1
    return Accepted, Solution

def SequenceGenerator(n):
    SeqSeed = ['1','2','3','4','3','2']
    IntSeq = []
    NewBuf = []
    SeedPointer = 0
    while(len(IntSeq)<n):
        NewBuf.append(SeqSeed[SeedPointer])
        Accepted, Solution = CheckNewSolution(NewBuf,IntSeq)
        if(Accepted):
            IntSeq.append(int(Solution)) 
            NewBuf = []
        SeedPointer += 1
        SeedPointer %= 6
    return IntSeq
```

```python
IntSeq = SequenceGenerator(11)
TotalSum = 0
for aux1 in range(len(IntSeq)):
    TotalSum = TotalSum + IntSeq[aux1]
print(TotalSum)
```
